```python
# segment data for assignment1
value_segment = pd.to_datetime('2016')
data_set['segment_fot_assignment'] = (pd.to_datetime(data_set['assignment1_submission']).
                                       apply(func = lambda x: 'early_finisher' if (x <= value_segment) else 'later_finisher'))

data_set['segment_fot_assignment1'] = np.where(pd.to_datetime(data_set['assignment1_submission']) <= value_segment, 'early_finisher', 'later_finisher')
```

```python
# excluid variables
data_set.drop(['segment_fot_assignment', 'segment_fot_assignment1'], axis = 1, inplace = True)
```

```python
# criate two subset data
early_finisher = data_set[pd.to_datetime(data_set['assignment1_submission']) < pd.to_datetime('2016')]
early_finisher.head()
```

```python
later_finisher = data_set[~data_set.index.isin(early_finisher.index)]
later_finisher.head()
```

```python
# mean of both groups
print("Mean of firt group: {} \nMean of second group: {}".format(early_finisher['assignment1_grade'].mean().round(2),
                                                                 later_finisher['assignment1_grade'].mean().round(2)))
```

**Hipótesis alternativa**: $\mu_1 \neq \mu_2, \ 1:\text{early} \ and \ 2:\text{later}$

```python
def t_student_assignment(df1, df2):
    
    # inicial values
    df1 = df1.iloc[:, 1:]
    df2 = df2.iloc[:, 1:]
    n_max = df1.shape[1]
    
    # loops
    result_test = []
    k = 0
    for i in range(0, n_max, 2):
        x = df1.iloc[:, i]
        y = df2.iloc[:, i]
        test_now = stats.ttest_ind(x, y)
        diff_mean = abs(x.mean() - y.mean())
        k += 1
        result_test.append((k, diff_mean, test_now[0], test_now[1]))
    
    # result
    result = pd.DataFrame(result_test, columns = ['assignment', 'diff_mean_abs', 'statistic', 'p_value'])
    return result
        
        
```

```python
t_student_assignment(early_finisher, later_finisher)
```

```python
# function
def t_student_assignment(df1, df2, alpha):
    
    # inicial values
    n_max = df1.shape[1]
    result_test = []
    k = 0
    
    # loops
    for i in range(0, n_max):
        k += 1
        x = df1.iloc[:, i]
        y = df2.iloc[:, i]
        
        diff_mean = abs(x.mean() - y.mean())
        test_now = stats.ttest_ind(x, y)
        
        if (test_now[1] >= alpha):
            text_conclusion = 'do not reject Ho'
        else:
            text_conclusion = 'do reject Ho'
            
        result_test.append((k, diff_mean, test_now[0], test_now[1], text_conclusion))
    
    # result
    result = pd.DataFrame(result_test, columns = ['assignment', 'diff_mean_abs', 'statistic', 'p_value', 'test_conclusion'])
    test_conclusion = result['test_conclusion'].value_counts(normalize = True) * 100
    print('rate do not reject Ho: {} % \nrate do reject Ho: {} %'.format(test_conclusion.loc['do not reject Ho'], test_conclusion.loc['do reject Ho']))
    result = result[result['test_conclusion'] == 'do reject Ho']
    return result
```

```python
# create data frame
data_set1 = pd.DataFrame(np.random.randint(1, 10, size = (100, 100)))
data_set2 = pd.DataFrame(np.random.randint(1, 10, size = (100, 100)))

# particular test 
t_student_assignment(data_set1, data_set2, 0.05)
```
